```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
comments = pd.read_csv('/content/filename.csv', error_bad_lines=False)
```

```python
comments.head()
```

```python
comments.iloc[:,10]
```

```python
#!pip install textblob
TextBlob("I am a madd girl").sentiment.polarity
```

```python
#This is to get sum of all missing values
comments.isna().sum()
```

```python
#To remove all missing value
comments.dropna(inplace=True)
```

```python
comments.head()
```

```python
polarity =[]

for i in comments['tweet']:
  polarity.append(TextBlob(i).sentiment.polarity)
```

```python
comments['polarity']= polarity
```

```python
comments.head()
```

```python
comments_positive = comments[comments['polarity']==1]
```

```python
comments_positive.head()
```

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
total_comments = ' '.join(comments_positive['tweet'])
```

```python
wordcloud=WordCloud(width=1000,height=500, stopwords=stopwords).generate(total_comments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

```python
comments_negative = comments[comments['polarity']==-1]
```

```python
comments_negative.head()
```

```python
total_comments = ' '.join(comments_negative['tweet'])
```

```python
wordcloud=WordCloud(width=1000,height=500, stopwords=stopwords).generate(total_comments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```
